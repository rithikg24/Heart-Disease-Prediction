```python
X = pd.get_dummies(df.drop('HeartDisease',axis=1))
```

```python
y=df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'n_estimators':[64,75,85,100,120,135,177]}
```

```python
gm_rfc=GridSearchCV(rfc,param_grid)
```

```python
gm_rfc.fit(X_train,y_train)
```

```python
gm_rfc.best_params_
```

```python
rfc_pred=gm_rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
plot_confusion_matrix(gm_rfc,X_test,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc_model=GradientBoostingClassifier()
```

```python
param_grid={'n_estimators':[50,100,150,200,250],
           'learning_rate':[0.1,0.05,0.2]}
```

```python
gm_gbc=GridSearchCV(gbc_model,param_grid)
```

```python
gm_gbc.fit(X_train,y_train)
```

```python
gm_gbc.best_params_
```

```python
gbc_pred=gm_gbc.predict(X_test)
```

```python
print(classification_report(y_test,gbc_pred))
```

```python
plot_confusion_matrix(gm_gbc,X_test,y_test)
```
